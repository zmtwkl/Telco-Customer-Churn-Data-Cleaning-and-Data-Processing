# Task 4 - Remove "(automatic)" from the paymentmethod column value and no spaces at the beginning or end

```python
df['paymentmethod']
```

```python
df['paymentmethod'] = df['paymentmethod'].str.replace('automatic','')
df['paymentmethod']
```

```python
df['paymentmethod'] = df['paymentmethod'].str.replace('()','', regex = False)
df['paymentmethod']
```

```python
df['paymentmethod'] = df['paymentmethod'].str.strip()
df['paymentmethod']
```

## 5b. Normalization for monthlycharges

```python
df['monthlycharges']
```

```python
from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler()
df['monthlycharges'] = min_max_scaler.fit_transform(df[['monthlycharges']])
df['monthlycharges']
```

## 5c. Standardization for totalcharges

```python
df['totalcharges']
```

```python
def isnumber(x):
  try:
    float(x)
    return True
  except:
    return False
df['isnumber'] = df['totalcharges'].apply(isnumber)
df['isnumber']
```

```python
index_notnumber =  df[~df['isnumber']].index
index_notnumber
```

```python
median = df[df['isnumber']]['totalcharges'].astype('float64').median()
median
```

```python
df.loc[index_notnumber,'totalcharges'] = median
```

```python
df.loc[index_notnumber]['totalcharges']
```

```python
df['totalcharges'].astype('float64')
```

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
df['totalcharges'] = std_scaler.fit_transform(df[['totalcharges']])
df['totalcharges']
```
